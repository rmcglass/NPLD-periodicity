```python
#Import packages
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft,rfft,ifft
import pandas as pd
import glob
```

```python
#Import csv files
csv_import = glob.glob("*zoom.csv")

column_names = ['depth', 'value']
csv_list = []
for i in csv_import:
    csv = pd.read_csv(i, names=column_names, skiprows=2)
    csv_list.append(csv)


    x = csv.depth.to_list()
    y = csv.value.to_list()
    
    peak = y.index(max(y[0:50]))
    surf = x[peak]
    print(surf)

    x_zoom = x[peak+5:]
    y_zoom = y[peak+5:]

    
    plt.figure(figsize = (12, 6))
    plt.subplot(121)
    plt.plot(x, y, 'r')
    plt.title(i)
    #zoom
    plt.subplot(122)
    plt.plot(x_zoom,y_zoom,'r')
    #plt.axvline(x=x_zoom[0]+5)
    plt.ylabel('Value')
    plt.xlabel('Depth (pixel)')

    plt.tight_layout()
    plt.show()
```

```python
fftVal = fft(y)
fftVal_zoom = fft(y_zoom)

#what is the x axis??

plt.figure(figsize = (12, 6))
plt.subplot(211)
plt.plot(x, np.abs(fftVal), 'b')

plt.subplot(212)
plt.plot(x_zoom, np.abs(fftVal_zoom), 'b')
plt.xlabel('?')
plt.ylabel('FFT Amplitude')

plt.show()
```
